5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data with correlated features
n_samples = 100
x = np.random.randn(n_samples)
y = 2 * x + np.random.randn(n_samples) * 0.5  # Correlated with x
z = 3 * y + np.random.randn(n_samples) * 0.2  # Correlated with y

# Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Compute the correlation matrix
corr_matrix = df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, fmt='.2f', cbar=True)

# Add a title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Generate the x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate the z values using the function z = sin(sqrt(x^2 + y^2))
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame for the data
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Set the layout of the plot
fig.update_layout(
    title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    coloraxis_colorbar=dict(
        title="Z Value"
    )
)

# Show the plot
fig.show()
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']

counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Prepare the output file (optional, can also display directly in Jupyter)
output_file("bar_chart.html")

# Create a figure object
p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")

# Add bars to the plot
p.vbar(x=fruits, top=counts, width=0.5, color="skyblue", legend_field="Fruits")

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"
p.xaxis.axis_label = "Fruit"
p.xaxis.major_label_orientation = "vertical"

# Show the plot
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file

# Generate the dataset (1000 random numbers from a standard normal distribution)
data_hist = np.random.randn(1000)

# Calculate histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Prepare the output file
output_file("histogram.html")

# Create a figure object for the histogram
p = figure(title="Histogram of Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           tools="pan, box_zoom, reset")

# Create the histogram using quad method to display bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)

x = np.linspace(0, 1, 10)

y = np.linspace(0, 1, 10)

xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.palettes import Viridis256
from bokeh.models import LinearColorMapper
from bokeh.transform import linear_cmap

# Generate dataset
data_heatmap = np.random.rand(10, 10)  # 10x10 matrix with random values
x = np.linspace(0, 1, 10)  # X-axis values
y = np.linspace(0, 1, 10)  # Y-axis values
xx, yy = np.meshgrid(x, y)  # Create meshgrid

# Prepare the output file
output_file("heatmap.html")

# Create a LinearColorMapper to map data values to colors using the Viridis palette
color_mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create a figure object for the heatmap
p = figure(title="Heatmap", 
           x_axis_label='X', 
           y_axis_label='Y', 
           tools="pan, box_zoom, reset")

# Add the heatmap to the figure using the 'image' method
p.image(image=[data_heatmap], 
        x=0, y=0, 
        dw=1, dh=1, 
        color_mapper=color_mapper)  # Apply the color mapper

# Customize the plot
p.x_range.range_padding = 0
p.y_range.range_padding = 0
p.grid.grid_line_color = None

# Show the plot
show(p)
```
